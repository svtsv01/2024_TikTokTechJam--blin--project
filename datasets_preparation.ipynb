```python
import pandas as pd
import math
```

```python
hodav = pd.read_csv('./Data/hate_and_offensive_davidson.csv', index_col=0)
civcom = pd.read_csv('./Data/civil_comments.csv')
```

```python
hodav.head()
```

```python
hodav = hodav[['tweet', 'hate_speech', 'offensive_language', 'neither', 'count']]
```

```python
def labeler_ho(row):
    if row['hate_speech'] >= math.ceil(row['count']/2):
        return 'hatespeech'
    elif row['offensive_language'] >= math.ceil(row['count']/2):
        return 'offensive'
    else:
        return 'normal'

hodav['label'] = hodav.apply(labeler_ho, axis = 1)
```

```python
hodav = hodav[['tweet', 'label']]
hodav = hodav.rename(columns={'tweet':'text'})
hodav
```

```python
hodav.label.value_counts()
```

```python
def labeler_cc(row):
    if row['toxicity'] >= 0.8:
        return 'hatespeech'
    elif (row['toxicity'] < 0.8) & (row['toxicity'] > 0.1):
        return 'offensive'
    else:
        return 'normal'

civcom['label'] = civcom.apply(labeler_cc, axis = 1)
civcom
```
