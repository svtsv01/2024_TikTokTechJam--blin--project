```python
import pandas as pd
import math, random
import nltk
import re
```

```python
#load the datasets
hodav = pd.read_csv('./Data_src/hate_and_offensive_davidson.csv', index_col=0)
civcom = pd.read_csv('./Data_src/civil_comments.csv')
hscom = pd.read_csv('./Data_src/hs_comments.csv')
```

```python
hodav.head()
```

```python
#pick the necessary fields
hodav = hodav[['tweet', 'hate_speech', 'offensive_language', 'neither', 'count']]
```

```python
#choose label if more then half of annotators picked it
def labeler_ho(row):
    if row['hate_speech'] >= math.ceil(row['count']/2):
        return 'hatespeech'
    elif row['offensive_language'] >= math.ceil(row['count']/2):
        return 'offensive'
    else:
        return 'normal'

hodav['label'] = hodav.apply(labeler_ho, axis = 1)
hodav.label.value_counts()
```

```python
#pick the necessary fields
hodav = hodav[['tweet', 'label']]
hodav = hodav.rename(columns={'tweet':'text'})
hodav
```

```python
#pick the necessary fields
civcom = civcom[['comment_text', 'toxicity']]
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
civcom[(civcom.toxicity > 0.7)].sort_values(by='toxicity', ascending=True).head(20)
```

```python
#apply labels depending on toxicity score
def labeler_cc(row):
    if row['toxicity'] >= 0.7:
        return 'hatespeech'
    elif (row['toxicity'] < 0.7) & (row['toxicity'] > 0):
        return 'offensive'
    else:
        return 'normal'

civcom['label'] = civcom.apply(labeler_cc, axis = 1)
civcom
```

```python
#pick necessary fields
civcom = civcom.rename(columns={'comment_text':'text'})
civcom = civcom[['text', 'label']]
```

```python
civcom.head()
```

```python
hodav.head()
```

```python
#combine the datasets
total_df = pd.concat([civcom, hscom, hodav]).reset_index().drop('index', axis = 1)
total_df['text'] = total_df['text'].str.replace('\n', ' ')
```

```python
#remove missing values
total_df = total_df.dropna()
```

```python
#import HateXplain dataset
import json

file = "./Data_src/HateXplain.json"
with open(file) as test_file:
    dict_test = json.load(test_file)
```

```python
#pick a random annotator's label

def annotatorSelect(value):
    return math.floor(random.uniform(0, 1)*len(value))

data = []

for key, value in dict_test.items():
    text = ' '.join(value['post_tokens'])
    label = value['annotators'][annotatorSelect(value['annotators'])]['label']
    data.append({'text': text, 'label': label})

df_test = pd.DataFrame(data)
```

```python
#create a numerical label
label_mapping = {'hatespeech': 2, 'offensive': 1, 'normal': 0}
df_test['label'] = df_test['label'].map(label_mapping)
total_df['label'] = total_df['label'].map(label_mapping)
```

```python
#remove every symbol that's not a letter
total_df['text'] = total_df.apply(lambda row: re.sub(r"[^\w]", " ", row['text']), axis = 1)
```

```python
#remove every symbol that's not a letter
df_test['text'] = df_test.apply(lambda row: re.sub(r"[^\w]", " ", row['text']), axis = 1)
```

```python
#Split the dataset
from sklearn.model_selection import train_test_split

df_train, df_val = train_test_split(total_df, random_state=420, test_size=0.15, stratify=total_df['label'])
```

```python
#assign fields to values
train_text, train_labels = df_train['text'], df_train['label']
val_text, val_labels = df_val['text'], df_val['label']
test_text, test_labels = df_test['text'], df_test['label']
```

```python
#save every field to file
train_text.to_csv('./Data_h/train_text.tsv', sep = '\t', index=False, header=False)
train_labels.to_csv('./Data_h/train_labels.tsv', sep = '\t', index=False, header=False)
val_text.to_csv('./Data_h/val_text.tsv', sep = '\t', index=False, header=False)
val_labels.to_csv('./Data_h/val_labels.tsv', sep = '\t', index=False, header=False)
test_text.to_csv('./Data_h/test_text.tsv', sep = '\t', index=False, header=False)
test_labels.to_csv('./Data_h/test_labels.tsv', sep = '\t', index=False, header=False) 
```

```python
#split the dataset
train_l, temp = train_test_split(df_test, stratify=df_test['label'], test_size=0.3)
test_l, val_l = train_test_split(temp, stratify=temp['label'], test_size=0.5)
```

```python
#save to file
train_l.to_csv('./Data_l/train_ll.csv', index=False)
test_l.to_csv('./Data_l/test_l.csv', index=False)
val_l.to_csv('./Data_l/val_l.csv', index=False)
```
